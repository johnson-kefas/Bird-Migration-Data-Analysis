```python
warbler_data = df[df['Species'] == 'Warbler']
```

```python
warbler_regions = warbler_data['Region'].value_counts()
print(warbler_regions)
```

```python
top_region = warbler_data['Region'].mode()[0]
print(f"Warbler migrates most frequently to: {top_region}")
```

```python
import matplotlib.pyplot as plt

warbler_regions.plot(kind='bar', color='green')
plt.title("Warbler Migrations by Region")
plt.xlabel("Region")
plt.ylabel("Number of Migrations")
plt.show()
```

```python
flock_data = df[df['Migrated_in_Flock'] == 'Yes']
```

```python
flock_by_species = df.groupby('Species')['Flock_Size'].agg(['mean', 'count', 'std'])
flock_by_species_sorted = flock_by_species.sort_values('mean', ascending=False)
print(flock_by_species_sorted)
```

```python
flock_by_species_sorted['mean'].plot(kind='bar', color='teal')
plt.title("Average Flock Size by Species")
plt.ylabel("Mean Flock Size")
plt.xlabel("Species")
plt.show()
```

```python
import seaborn as sns

sns.boxplot(data=df, x='Species', y='Flock_Size')
plt.title("Flock Size Distribution by Species")
plt.xticks(rotation=45)
plt.show()
```

```python
from scipy.stats import f_oneway

goose_flocks = df[df['Species'] == 'Goose']['Flock_Size']
warbler_flocks = df[df['Species'] == 'Warbler']['Flock_Size']

f_stat, p_value = f_oneway(goose_flocks, warbler_flocks)
print(f"P-value: {p_value:.4f}")  # Significant if p < 0.05
```
